```python
def passenger_arrival_at_tick(row):
    
    def passenger_arrival(tick, source_id, destination_id, passenger_count):
        
        for i in range(passenger_count): #TODO change id
            self.bus_stops[source_id].add_waiting_passenger(new Passenger(self.passenger_count , source_id, destination_id, tick))
            self.passenger_count +=1
            
    tick = row['Ticks']  # When the passengers arrive (every 15 ticks)
    source_name = row['FROM'] # source station
    destination_names_counts = row[ams][row[ams] != 0] # [(destination, num of passengers)]
    
    # For each destination-passenger tuple
    for destination_name, passenger_count in destination_names_counts.items():
        source_id = Controller.amsterdam_bus_stops_names[source_name]
        destination_id = Controller.amsterdam_bus_stops_names[destination_name]
        
        passenger_arrival_action = Action(tick, partial(passenger_arrival, tick, source_id, destination_id, passenger_count))
        
        self.actions.push(passenger_arrival_action)
    
df.apply(passenger_arrival_at_tick, axis=1)

```
